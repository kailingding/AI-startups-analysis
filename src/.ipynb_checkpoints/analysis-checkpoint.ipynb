```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import random
pallet = ['#D61A46', '#FDDC22', '#236AB9', '#342E09']
```

```python
# read csv file and intialize AI startups dataset 
df = pd.read_csv("startups.csv").drop(columns=["link", "website"])
df.head(200)
```

```python
# convert raised to double
import re
# convert currency to numerics
def currency(s):
    try:
        return float(''.join(re.findall(r'\d', s)))
    except:
        return np.nan

df.raised = df.raised.apply(currency)
```

```python
from datetime import datetime
def process_date(date):
    date = datetime.strptime(date, "%b %y")
    return date
```

```python
# most used programming language among AI startups
programming = df.groupby(['tech']).count().name
fig = plt.figure(figsize=[20,4])
plt.title('Tech', fontsize=25)
programming.plot(kind='bar', color=random.choice(pallet), fontsize=15, rot=0)
plt.xlabel('')
plt.show()
```
